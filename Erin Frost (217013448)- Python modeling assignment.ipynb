# <u>The development of the novel phenotype of citrate metabolism, in populations of *Escherichia coli* (from the LTEE), in relation to favoured mutations </u>

## <u>Introduction</u>

The Long-Term Evolution Experiment (LTTE), that was conducted by Lenski and team  in 1988, was orignially based on 12 populations of *Escherichia coli* that are from the same ancestral strain (Lenski et al., 1991). This experiment served the purpose of studying the evolutionary changes within these populations of bacteria, that have extremely short generation intervals, through the accumulation of various mutations. Due to the very small generation intervals of *E.coli*, sudden changes in the growth patterns of this bacteria were recorded, and therefore the adaption potential of the bacteria to their environment was determined (Lenski, 2017). The adaption potential of these laboratory-based populations, could therefore be compared to natural populations, as to whether the rate of evolution would be severly different or not. According to Rozen and Lenski (2000), one of the major objectives of this experiment was to determine whether the 12 populations of the same lineage would diverge from each other, due to accumulation of different mutations as well as different mutational rates across generations. The 12 populations were subjected to specific environmental conditions, whereby glucose in the growth medium was a limiting factor and citrate was incorporated, which was aimed to accelerate selection pressures (Vasi et al., 1994). Due to the fact that glucose was a limiting factor as a carbon source, the bacteria had to adapt to an unfavourable environment in order to survive, and therefore the spontaneous development of citrate metabolism in the Ara-3 population was proposed to have come of this (Van Hofwegen et al., 2016).

Citrate metabolism plays an important role in fermentation processes that involve anaerobic micro-organisms and therefore cannot be exlpoited by aerobic micro-organisms - such as *E.coli* (Leiby et al., 2012). E.coli cannot metabolize citrate under normal aerobic conditions, as the bacteria lacks expression of an existing gene that codes for a citrate transporter (*CitT*), which is only activated in anaerobic conditions (Roth & Maisnier-Patin, 2016). According to Blount et al. (2012), the evolution of citrate metabolism occured through three phases (that contains one or more mutations): potentiation, actualization and refinement. The key mutations occur in the actualization phase, whereby the Cit+ phenotype manifests due to the duplication of the *rnk-citG* gene region in the *E.coli* genome, which contains the citrate transporter (*CitT*) gene. One of the two copies of the *CitT* gene region is converted into a metabolic pathway, that allows for aerobic respiration, and can therefore be exploited by *E.coli* to obtain energy. The refinement phase involves a defining mutation, deemed as the *dctA* gene mutation, that improves efficiency of the metabolism of citrate by *E.coli* and creates clones with the *Cit++* phenotype (Leon et al., 2018). According to Blount et al. (2008), the first eveidence of these *Cit+* variant clones evolved in the Ara-3 population, around 31,500 generations and the overall population fitness increased dramatically.

The objectives of this modeling project was to determine whether the number of total mutations increased aorund 31,500 generations of the Ara-3 population increased when the phenotypic trait of citrate metabolism (*Cit+*) evolved and to determine what types of mutations were favoured around 31,500 generations of the Ara-3 population. The hypothesis of this experiment was to ask whether favoured mutations are directly related to the development of a novel function, in this case, the evolution of citrate metabolism in the Ara-3 *E.coli* population.

## <u> Discussion</u>

The ability of a population of *E.coli*, to spontaneously develop the novel function of aerobic citrate-metoblism, was a major innovation in the survival of the population and initiated the evolution of the *E.coli* lineage (Turner et al., 2015). According to Blount et al. (2008), this novel function developed in the Ara-3 population, where a citrate-metabolic (*Cit+*) variant evolved and was able to aerobically metabolize citrate in order to survive. As seen in Figure 1, after 30,000 generations the total number of mutations increased rapidly, until 40,000 generations was reached and a severe decline in the number of total mutations was obsevered. As Bahar (2018) points out, over many generations in a rapidly-reproducing bacterial population (such as *E.coli*), the accumulation of mutations (contributing top the baseline mutation rate) is expected in order for a population with very small generation intervals to adapt and survive in a rapidly-changing environment. Although this was taken into consideration, the rapid increase in the number of mutations was expected after 30,000 generations of the Ara-3 population, as a new phenotype evolved in the population and the evolving lineage contained new mutations that lead to the novel phenotype that the ancestral *E.coli* strain did not contain. Bahar (2018) also made the conclusion that the rapid increase of the total number of mutations after 30,000 generations, was very slow comapared to the baseline mutation rate (that was determined in previous studies using the LTEE), and therefore indicates that the *Cit+* phenotype evolved from a complex mechanism rather than a single point mutation. This statement correlates with Blount et al. (2012), whereby the process of activating the citrate transporter (*CitT*) in the genome goes through three phases, and the "potentiating" phase (initial phase) would contain multiple mutations that leads to the activation of the citrate transporter (*CitT*). In Figure 1, the total number of mutations rapidly decreases at 40,000 generations (which was after the proposed time that the aerobic-metabolic phenotype evolved in the population), which may be due to the expression of some 'antimutator' genes or the epistatic interaction between the accumulated genes (Raeside et al., 2014).

In Figure 2 and Figure 3, the favoured mutations after the evolution of the citrate variants (*Cit+*) at 31,500 generations that were determined was the base substitution mutations, that was composed mainly of nonsynonymous substitutions, synonymous substitutions and intergenic substitutions. Blount et al. (2012) indicates that the final expression of the citrate transporter (*CitT*), to allow the phenotype of citrate-metabolism to be expressed, was due to the duplication of the *rnk-citG* gene region. As the population reaches 40,000 generations, it was expected that the number of large duplication mutations would increase rapidly, as the population evolves to contain the *Cit+* phenotype. This was not the case in the Ara-3 *E.coli* population, as the evolution of the expression of the *Cit+* phenotype was deemed as a three-phased process, that has the potential to contain numerous favourable mutations in order to exploit the full expression of the citrate transporter (*CitT*). Hall (1982) expresses the fact that the genes involved in the "potentiating" step of the expression of the citrate-transporter (*CitT*) was not determined and still remains unknown. Therefore, the rapid increase of base substitution mutations (that include nonsynonymous, synonymous and intergenic substitution mutations) indicates the favoured mutations within the population, that are favoured due to the fact that the mutations form part of the multi-phased process of the expression of the citrate (*Cit+*) phenotype.

Therefore, the hypothesis that was proposed was accepted. The favoured mutations that were determined within the population may not be the actual mutation that causes the expression of the citrate phenotype (*Cit+*) - which is due to a duplication mutation in the variants. But, the favoured base substitution mutations have a very high likehood of being directly-related to the development of the novel citrate-metabolism ability of these population variants - within the "potentiating" phase of the phenotype expression.
